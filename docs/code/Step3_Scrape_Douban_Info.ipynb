```python
from bs4 import BeautifulSoup
import requests

from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

import pandas as pd
import re
```

```python
df = pd.read_csv("all_movie_info.csv")
unique_movie = df.drop_duplicates(subset=['movie_name', 'movie_director','movie_year'])
moviename_list = unique_movie.movie_name.tolist()
```

```python
hrefs = []
for movie_name in moviename_list:
    one_movie  = {}
    one_movie['name'] = movie_name
    name_string = movie_name.replace(" ","+").lower()
    href = "https://www.douban.com/search?q=" + name_string
    one_movie['douban_href'] = href
    hrefs.append(one_movie)
len(hrefs)
```

```python
douban  = []
for href in hrefs:
    try:
        driver = webdriver.Chrome(ChromeDriverManager().install())
        driver.get(href['douban_href'])
        soup_doc = BeautifulSoup(driver.page_source)
        one_movie = {}
        one_movie['name'] = href['name']
        one_movie['douban_rate'] = soup_doc.find(class_='rating_nums').text
        douban_ppl = soup_doc.find(class_='rating_nums').next.next.next.text
        one_movie['douban_ppl'] = re.findall(r'[0-9]+',douban_ppl)[0]
        douban.append(one_movie)
        driver.close()
    except:
        one_movie['douban_rate'] = "NaN"
        one_movie['douban_ppl'] = "NaN"
        douban.append(one_movie)
```

```python
pd.DataFrame(douban).to_csv('douban_info.csv',index=False)
```
