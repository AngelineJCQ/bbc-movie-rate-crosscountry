```python
import pandas as pd
```

```python
movie = pd.read_csv("all_movie_info.csv")
director = pd.read_csv("birthplace.csv")
douban = pd.read_csv("douban_info.csv")
imdb = pd.read_csv("imdb_info.csv")
```

```python
add_director = pd.merge(movie, director, how='left', left_on="movie_director", right_on="name").drop_duplicates()
add_douban = pd.merge(add_director, douban, how='left', left_on="movie_name", right_on="name").drop_duplicates()
add_imdb = pd.merge(add_douban, imdb, how='left', left_on="movie_name", right_on="name").drop_duplicates()
df = add_imdb.drop(['name_x', 'lat', 'lng', 'name_y','name'], axis=1)

#Turn M and K in the imdb_ppl column to integer
df['imdb_ppl1'] = df.imdb_ppl.replace(r'[KM]+$', '', regex=True).astype(float) * \
    df.imdb_ppl.str.extract(r'[\d\.]+([KM]+)', expand=False).fillna(1).replace(['K','M'], [10**3, 10**6]).astype(int)
df2 = df.drop(['imdb_ppl'], axis=1)
df2['rate_dif'] =  df2['douban_rate'] -  df2['imdb_rate']
df2['ppl_dif'] =  df2['douban_ppl'] -  df2['imdb_ppl1']
df2
```

```python
df2.to_csv("final_detail_dataframe.csv",index=False)
```
