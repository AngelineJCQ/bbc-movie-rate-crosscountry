## Do some import

```python
from bs4 import BeautifulSoup
import requests

from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

import pandas as pd
import re
```

## Scrape IMDB data

```python
pages = [1,2]
all_movie_info = []
for page in pages:
    link = 'https://www.imdb.com/list/ls025871744/?sort=list_order,asc&st_dt=&mode=detail&page=' + str(page)
    raw_html = requests.get(link).content
    soup_doc = BeautifulSoup(raw_html, "html.parser")
    list_div = soup_doc.find(class_ = 'lister-list').find_all(class_ = 'lister-item mode-detail')
    for div in list_div:
        one_movie = {}
        one_movie['name'] = div.find_all('a')[1].string
        one_movie['director'] = div.find_all(class_ = 'text-muted text-small')[1].find('a').string
        one_movie['link']  = div.find_all('a')[1]['href']
        year = div.find(class_ = 'lister-item-year text-muted unbold').string
        one_movie['year'] = re.findall(r'[0-9]+', year)[0]
        length = div.find(class_ = 'runtime').string
        one_movie['runtime'] = re.findall(r'[0-9]+', length)[0]
        one_movie['genre']  = div.find(class_ = 'genre').string.strip()
        one_movie['a_rate'] = div.find(class_ = 'ipl-rating-star__rating').string.strip()
        one_movie['a_ppl'] = div.find_all('span', {"name":"nv"})[0].string
        try:
            one_movie['gross'] = div.find_all('span', {"name":"nv"})[1].string
        except:
            one_movie['gross'] = "NaN"
        all_movie_info.append(one_movie)
all_movie_info
```

## Scrape Douban data

```python
start_number = [0, 25, 50, 75, 100]
links = []
for number in start_number:
    link = 'https://www.douban.com/doulist/45038321/?start='+ str(number) +'&sort=time&playable=0&sub_type='
    links.append(link)

driver = webdriver.Chrome(ChromeDriverManager().install())
all_movie_info2 = []
for page in links:
    driver.get(page)
    soup_doc = BeautifulSoup(driver.page_source)
    all_movies = soup_doc.find_all(class_ = 'doulist-item')
    for movie in all_movies:
        movie_info = {}
        ppl_info = movie.find(class_="rating_nums").next.next.next.string
        movie_info['name'] = movie.find('div', class_ = 'title').text.strip()
        movie_info['c_rate'] = movie.find(class_="rating_nums").string
        movie_info['c_ppl'] = re.findall(r'[0-9]+', ppl_info)[0]
        all_movie_info2.append(movie_info)
all_movie_info2
```

## Transfer them into one datafrme

```python
df1 = pd.DataFrame(all_movie_info)
df2 = pd.DataFrame(all_movie_info2)
df = pd.concat([df1, df2], axis=1)
```

```python
df
```

## Clean the data

```python
df['runtime'] = df['runtime'].astype('int')
df['a_rate'] = df['a_rate'].astype('float')
df['c_rate'] = df['c_rate'].astype('float')
df['c_ppl'] = df['c_ppl'].astype('int')
df['a_ppl'] = df['a_ppl'].str.replace(',','').astype('int')
df['gross'] = df['gross'].str.replace('$','')
df['gross'] = df['gross'].str.replace('M','').astype('float')
df.dtypes
```

```python
df['df_rate'] = df['a_rate'] - df['c_rate']
df['df_ppl'] = df['a_ppl'] - df['c_ppl']
df.to_csv('rate_ppl.csv',index=False)
```
